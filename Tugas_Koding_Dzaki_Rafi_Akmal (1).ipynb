**TensorFlow**

```python
import tensorflow as tf #mengimport tensorflow

x_tf = tf.constant(10) #menyimpan nilai tensorflow x
x_tf
x_tf.dtype             #menampilkan tipe data dari tensorflow x
```

```python
y_tf = tf.constant(5) #menyimpan nilai tensorflow y
y_tf
y_tf.dtype            #menampilkan tipe data dari tensorflow
```

```python
x_tf + y_tf #menjumlahkan hasil dari tensorflow
```

```python
x_float_tf = tf.constant(3.5) ##menampilkan tipe data dari x_float_tf
x_float_tf
x_float_tf.dtype
```

```python
x_float_tf + tf.cast(y_tf, tf.float32) #menjumlahkan hasil
```

**Pytorch**

```python
import torch #mengimport torch

x = torch.tensor(10) #menyimpan nilai x
x
x.dtype #menampilkan tipe data
```

```python
y = torch.tensor(5) ##menyimpan nilai y
y
y.dtype #menampilkan tipe data
```

```python
x + y #menjumlahkan x+y dengan torch
```

```python
x_float = torch.tensor(3.5)
x_float
x_float.dtype
```

```python
x_float + y.float()
```
