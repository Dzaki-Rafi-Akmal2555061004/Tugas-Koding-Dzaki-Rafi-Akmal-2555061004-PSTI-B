Vectors and Vector Transposition


```python
import numpy as np
```

Norms and Unit Vectors


```python
import numpy as np
```

```python
# vector x ukuran 3x1
x = np.array([[1],
              [2],
              [3]])
x
```

```python
(1**2 + 2**2 + 3**2)**(1/2)
```

```python
# L2 norm dari x
norm_x = np.linalg.norm(x)
norm_x
```

```python
torch.norm(torch.tensor([1, 2, 3.]))
```

```python
# vector v dengan elemen (3,4)
v = np.array([3, 4])
v
```

```python
# L2 norm dari v
norm_v = np.linalg.norm(v)
norm_v
```

```python
# menampilkan unit vector dari v
unit_v = v / norm_v
unit_v
```

```python
# menghitung panjang unit vector
panjang_unit_v = np.linalg.norm(unit_v)
panjang_unit_v
```

Basis, Orthogonal, and Orthonormal Vectors

```python
import numpy as np
```

```python
# membuat vector i dan j
i = np.array([1, 0])
j = np.array([0, 1])
i,j
```

```python
# membuktikan dot product
dot_ij = np.dot(i, j)
dot_ij
```

```python
# membuat vector v dan w
v = np.array([1, 2])
w = np.array([2, -1])
v,w
```

```python
# membuktikan dot product
dot_vw = np.dot(v, w)
dot_vw
```

Matrix Tensors

```python
import numpy as np
```

```python
# membuat matrix X ukuran 2x3
X = np.array([[1, 2, 3],
              [4, 5, 6]])
X
```

```python
# ukuran matrix
ukuran_X = X.shape
ukuran_X
```

```python
# kolom pertama
kolom_pertama = X[:, 0]
kolom_pertama
```

```python
# baris kedua
baris_kedua = X[1, :]
baris_kedua
```
